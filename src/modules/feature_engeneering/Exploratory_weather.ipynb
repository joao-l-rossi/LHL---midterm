```python
columns_to_keep = ["arr_delay","op_unique_carrier", "tail_num", "origin", "dest", "crs_elapsed_time", "distance",
                  "dep_Rain","dep_Fog","dep_Snow","dep_Cold", "dep_Storm", "dep_Hail", "dep_Precipitation",
                   "arr_Rain", "arr_Fog", "arr_Snow", "arr_Cold", "arr_Storm", "arr_Hail", "arr_Precipitation", 
                  "crs_dep_time", "crs_arr_time", "dup"]
```

```python
df_train = df_train[columns_to_keep]
```

```python
numeric_df = df_train[df_train.columns[df_train.dtypes != object]]
```

```python
ax = sns.boxplot(x="dup", y="arr_delay", data=df_train)
ax.set_ylim([-100, 150]) 
```

```python
temp = min(df_train["arr_delay"].dropna())
df_train["log_delay"] = (df_train["arr_delay"] - temp + 1).apply(np.log)
```

```python
arrival = df_train[["arr_delay", "arr_Rain"]]
arrival["Event"] = "Arrival"
arrival.rename(columns = {"arr_Rain":"Rain"},inplace=True)
departure = df_train[["arr_delay", "dep_Rain"]]
departure["Event"] = "Departure"
departure.rename(columns = {"dep_Rain":"Rain"}, inplace=True)
```

```python
data_temp = pd.concat([arrival,departure], axis=0)
```

```python
data_temp
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.collections import PathCollection


ax = sns.violinplot(x="Rain", y="arr_delay", data=data_temp,
                    color=".8", hue = "Event", palette='RdYlGn',
                    order = ["NW", "Light", "Moderate", "Heavy"],split = True)

ax.set_ylabel("Arrival Delay")
ax.set_xlabel("Rain Intensity")

for artist in ax.lines:
    artist.set_zorder(10)
for artist in ax.findobj(PathCollection):
    artist.set_zorder(11)
plt.legend(loc='upper left')
#sns.stripplot(x="dep_Rain", y="arr_delay", data=df_train, jitter=0.08, ax=ax, size = 4, alpha=0.1)
ax.set_ylim([-110, 250]) 
plt.show()
```
