# **Task 7**: Test the hypothesis whether planes fly faster when there is the departure delay?
#
### h0= Planes do not fly faster when there is a departure delay

```python
#this cell block deals with data where there was a departure delay

#SQL:
'''select crs_elapsed_time, actual_elapsed_time, (crs_elapsed_time-actual_elapsed_time) faster_flying	
from flights f 
where dep_delay >0
and arr_delay is not null
;'''

#read csv where the the airplane is late on departure 
df1=pd.read_csv('q7_delay.csv')

#Computing the ratio of times the pilot flew faster when the plane was late on departure
df1_faster=df1.faster_flying>0 # if faster_flying is positive, that means the pilot flew faster
ratio=round(len(df1[df1_faster]) / len(df1)*100,2)
print(ratio)
df1
```

```python
x=df1['faster_flying']
```

```python
sns.boxplot(data=x)
```

```python
x.describe()
```

```python
#Looking at the departure delays where the pilots flew 100 mins slower than their predicted flight time 

#sql
'''select *	
from flights f 
where dep_delay > 0
and arr_delay is not null
and (crs_elapsed_time-actual_elapsed_time)<-100
'''

df=pd.read_csv('q7.lessthan100.csv')
df[['carrier_delay','weather_delay','nas_delay','security_delay','late_aircraft_delay']].describe() #look at NAS median and mean
```

```python
#this cell deals with instances where the on departure the flight was on-schedule or early

#SQL:
'''select crs_elapsed_time, actual_elapsed_time, (crs_elapsed_time-actual_elapsed_time) faster_flying	
from flights f 
where dep_delay <= 0
and arr_delay is not null
;'''

#read csv where the the airplane is on time or early
df2=pd.read_csv('q7_ontime_early.csv')

#Computing the ratio of times the pilot flew faster 
df2_faster=df2.faster_flying>0 # if faster_flying is positive, that means the pilot flew faster
ratio2=round(len(df2[df2_faster]) / len(df2) * 100,2)
print(ratio2)
df2
```
